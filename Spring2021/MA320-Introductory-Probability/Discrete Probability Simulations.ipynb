```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def toss(n):
    return [int(np.floor(x)) if x < 0.5 else int(np.ceil(x)) for x in np.random.rand(n)]
```

```python
results = []
for i in range(1, 1000):
    tosses = toss(i)
    heads = tosses.count(1)
    proportion_heads = heads / len(tosses)
    results.append(proportion_heads - 0.5)
```

```python
results = np.array(results)
y_ax = np.linspace(1, 1000, 999)

print(results.shape, y_ax.shape)
plt.scatter(results, y_ax)
```

```python
# election simulation
pop_size = 100000
population = [int(np.floor(x)) if x < 0.48 else int(np.ceil(x)) for x in np.random.rand(pop_size)]# create a population with given distribution
print(population.count(0)/pop_size)
```

```python
def simulate_poll(sample_size, population):
    return np.random.choice(population, sample_size)
```

```python
# Actual Simulation
sample_size = 10000
r_props = np.zeros(1000)
d_props = np.zeros(1000)
for i in range(1000):
    simulation = simulate_poll(sample_size, population)
    d_votes = np.count_nonzero(simulation)
    r_votes = sample_size - d_votes
    r_props[i] = r_votes / sample_size
    d_props[i] = d_votes / sample_size
```

```python
y = np.linspace(0, 1000, 1000)
plt.style.use('fivethirtyeight')
plt.scatter(y, d_props, color='blue')
plt.scatter(y, r_props, color='red')
plt.axhline(y=0.48, color='g', linestyle='--')
plt.axhline(y=0.52, color='g', linestyle='--')
plt.xlabel('Epoch')
plt.ylabel('Proportion of Sample Size')
```
