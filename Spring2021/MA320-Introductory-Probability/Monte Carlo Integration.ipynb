```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def rand_xy():
    return np.random.rand(1,2)
def simulate(iterations):
    results = np.zeros((iterations, 2))
    for i in range(iterations):
        results[i, :] = rand_xy()
    return results
```

```python
iterations = 2500
results = simulate(iterations)
results_itm = []
for coordinate in results:
    if np.sqrt(coordinate[0]) > coordinate[1]:
        results_itm.append((coordinate[0], coordinate[1]))
results_in_area = np.array(results_itm)
```

```python
plt.scatter(results_in_area[:, 0], results_in_area[:, 1])
x = np.linspace(0, 1, iterations)
y = np.sqrt(x)
plt.plot(x, y, 'r')
print("Best integral approximation of sqrt(x) from 0 to 1:", results_in_area.shape[0] / iterations)
```
