```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# load data
data = pd.read_csv('House_Data.csv') #data stored in a structure DataFrame
print(data.shape)
data.head() #print the first 5 rows
#data.describe() #basic statistical measures
print(np.any(data.isnull()==True))
```

```python
#basic plotting
fig=plt.figure()
plt.scatter(data['sqft_living'],data['price']) #Dataframe indexing: loc (indexed by column name) iloc(idexed by number)
plt.show()
size=fig.get_size_inches()
print(size)

plt.figure(figsize=(6,72))
for i in range(18):
    plt.subplot(18,1,i+1)
    plt.scatter(data.iloc[:,i+3],data.iloc[:,2])
    plt.title('Price vs.' + data.columns[i+3])
plt.show()
```

```python
# converting dataframe to numpy array
#X = data.iloc[:,3:]
X = data.loc[:,['bedrooms','bathrooms','sqft_living','grade','sqft_above','sqft_living15']]
Y = data.iloc[:,2]

X = X.to_numpy()
Y = Y.to_numpy() #convert to numpy array
print(X.shape, Y.shape)
```

```python
#data normalization, to make the contribution from factors to prediciton equal, to [0 1], to [0.05 0.95], (X-Xmin)/(Xmax-Xmin)*0.9+0.05
for i in range(X.shape[1]):
    data_ = X[:,i]
    X[:,i] = (data_-np.amin(data_))/(np.amax(data_)-np.amin(data_))*0.9+0.05
const = np.ones((X.shape[0],1))

X = np.concatenate((const,X),axis=1) #axis = 0 group vertically, axis =1 group horizontally
print(X[0:5,:])
```

```python
# data splitting, build training and testing data; 70% for training, 30% for testing
m = 0.7 * X.shape[0]
m = np.ceil(m) # round the number, or using floor fuction
m = int(m)
X_Train = X[:m,:]
Y_Train = Y[:m]
X_Test = X[m:,:]
Y_Test = Y[m:]
print(X_Train.shape, Y_Train.shape)
```

```python
# define the loss function
def loss_function(X,Y,B):
    J = np.sum((X.dot(B)-Y)**2)/(2*X.shape[0])
    return J
```

```python
# create multiple regression with gradient descent
def gradient_descent(X,Y,B,Iterations,alpha):
    loss_history = np.zeros(Iterations)
    for i in range(Iterations):
        gradient = X.T.dot((X.dot(B)-Y))/X.shape[0] #X.T(XB-Y)/N
        B = B - alpha*gradient
        loss_history[i] = loss_function(X,Y,B)
    return B, loss_history
```

```python
B = np.random.random(X_Train.shape[1])
Iterations = 20000
alpha = 0.01
B_new, loss_history = gradient_descent(X_Train,Y_Train,B,Iterations,alpha)

plt.plot(np.arange(Iterations), loss_history)
```

```python
# performance evaluation using testing data
Y_Test_prediction = X_Test.dot(B_new) #6483x19 19x1 ->6483x1
plt.scatter(Y_Test,Y_Test_prediction)
plt.xlim((0,2500000))
plt.ylim((0,2500000))
```

```python
# performance evaluation using RMSE, R2
def RMSE(Y_actual,Y_prediction):
    rmse=np.sqrt(np.sum((Y_actual-Y_prediction)**2)/len(Y_actual))
    return rmse
RMSE(Y_Test,Y_Test_prediction)
#print(Y_Test.shape, Y_Test_prediction.shape)

def R2(Y_actual,Y_prediction):
    sst=np.sum((Y_actual-Y_actual.mean())**2)
    ssr=np.sum((Y_actual-Y_prediction)**2)
    r2 = 1-(ssr/sst)
    return r2
R2(Y_Test,Y_Test_prediction)
```
