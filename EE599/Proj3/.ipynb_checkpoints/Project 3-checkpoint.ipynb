## Multiple Regression and Scikit Learn Models

```python
# Import Necessary Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn-pastel')
np.random.seed(2)
```

```python
# Load the Data
data = pd.read_csv('data.csv') # Read The CSV
print(data.shape)
data.head() # Print First 5 Rows
```

```python
# Data Analysis
plt.figure(figsize=(6, 48))
for i in range(8):
    plt.subplot(8, 1, i+1)
    plt.scatter(data.iloc[:,i+2],data.iloc[:,1])
    plt.title('Price vs. ' + data.columns[i+2])
plt.show()
```

```python
# Data Frame -> Numpy Array
X = data.loc[:,['T_Mold', 'T_Melt']].to_numpy() # Sensing Variables
Y = data.loc[:,['Part thickness']].to_numpy().ravel() # Because of how we're accessing this column we need to flatten the array
print(X.shape, Y.shape)
print(X[0:5, :])
```

```python
# Normalize the data
for i in range(X.shape[1]):
    data_ = X[:,i]
    X[:,i] = (data_-np.amin(data_))/(np.amax(data_)-np.amin(data_))*0.9+0.05
const = np.ones((X.shape[0],1))

X = np.concatenate((const,X),axis=1) 
print(X[0:5, :])
```

```python
# Split the Data, 70% for Training, 30% of Testing
np.random.shuffle(X)
m = int(np.ceil(0.7 * X.shape[0])) # Ceil() the number then convert to integer
x_train = X[:m,:]
y_train = Y[:m]
x_test = X[m:, :]
y_test = Y[m:]

print(x_train.shape, y_train.shape)
```

```python
# Define the loss function
def loss_function(X, Y, B):
    J = np.sum((X.dot(B)-Y)**2)/(2*X.shape[0])
    return J
```

```python
# Define a Multiple Regression using the Gradient Descent algorithm
def gradient_descent(X,Y,B,Iterations,alpha):
    loss_history = np.zeros(Iterations)
    for i in range(Iterations):
        gradient = X.T.dot((X.dot(B)-Y))/X.shape[0]
        B = B - alpha*gradient
        loss_history[i] = loss_function(X,Y,B)
    return B, loss_history
```

```python
# Find an Optimal Learning Rate
B = np.random.random(x_train.shape[1])
iterations = 20000

plt.figure(figsize=(6, 24)) # Make the analysis viewable 
for i in range(5):
    plt.subplot(5, 1, i+1)
    alpha = 1 / np.power(10, i+2) # create alpha value by 1 * 10 ^ -i
    _, loss_history = gradient_descent(x_train, y_train, B, iterations, alpha) # Test this alpha, throw out first result til later
    plt.title("Test Alpha: " + str(alpha))
    plt.plot(np.arange(iterations), loss_history) # Plot test results
plt.show()
```

## Analysis of Learning Rate
From our little data analysis above it can be seen that the most optimal learning rate is when alpha is equal to 0.001. Anything above this would result in the possibility we may skip over our optimal weighting. While anything below it wil lresult in many iterations required to reach those values.

```python
# re run test with desire learning rate
alpha = 0.001
B_new, loss_history = gradient_descent(x_train, y_train, B, iterations, alpha)
```

```python
# performance evaluation using test data
y_test_prediction = x_test.dot(B_new)
plt.scatter(y_test, y_test_prediction)
plt.xlabel("Actual Part Thickness")
plt.ylabel("Predicted Part Thickness")
```

```python
# performance evaluation using RMSE
def RMSE(y_actual, y_prediction):
    return np.sqrt(np.sum((y_prediction-y_actual)**2)/len(y_actual))
RMSE(y_test, y_test_prediction)
```

## 3 Layer Neural Network Perceptron

### Prerequisites

```python
# Import Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
np.random.seed(2)
```

```python
# Load Data
data = pd.read_csv('data.csv')

# Convert to NP Array
X = data.loc[:,['T_Mold', 'T_Melt']].to_numpy() # Sensing Variables
Y = data.loc[:,['Part thickness']].to_numpy().ravel() # Because of how we're accessing this column we need to flatten the array
print(X.shape, Y.shape)
```

```python
# data normalization
np.random.shuffle(X)
x_norm = np.zeros((X.shape[0], X.shape[1]))
for i in range(X.shape[1]):
    data_ = X[:, i]
    x_norm[:, i] = (data_-np.amin(data_))/(np.amax(data_)-np.amin(data_))

y_min = np.min(Y)
y_max = np.max(Y)
y_norm = (Y-y_min)/(y_max-y_min)
```

```python
m = int(np.ceil(0.7 * X.shape[0]))

x_train = x_norm[:m, :]
y_train = y_norm[:m]
x_test = x_norm[m:, :]
y_test = y_norm[m:]
y_train = np.reshape(y_train, (len(y_train),1))
y_test = np.reshape(y_test, (len(y_test),1))
print(x_test.shape, y_test.shape)
```

```python
# Sigmoid function and its derivative
def sigmoid(x):
    output = 1/(1+np.exp(-x))
    return output
def sigmoid_der(output):
    return output*(1-output)
```

```python
# 3 Layer Perceptron w/ Feed-forward and back-propagation
class NeuralNetwork():
    def __init__(self,x,y,layer_numbers,learning_rate,iterations): #layer_numbers: [6, 10, 1], role of initialization funcition is make local variables as global variables
        self.input = x
        self.y = y
        self.layer_numbers = layer_numbers
        self.learning_rate = learning_rate
        self.iterations = iterations
        #define network structure
        self.W0 = np.random.rand(self.layer_numbers[0], self.layer_numbers[1]) #6*10
        self.W1 = np.random.rand(self.layer_numbers[1], self.layer_numbers[2]) #10*1
        #for the purpose of learning curve plot
        self.iteration = []
        self.error_history = []
    
    def forward(self): #with self as input variable, we can call all variables that starts with self.
        self.hidden_output = sigmoid(np.dot(self.input, self.W0)) #sigmoid(w0*x)
        self.output = sigmoid(np.dot(self.hidden_output, self.W1)) #sigmoid(w1*h)
    
    def backpropagation(self):
        self.loss=np.sum((self.output-self.y)**2)/2
        d_W1 = np.dot(self.hidden_output.T, (self.output-self.y)*sigmoid_der(self.output))  #hT*(y-yactual)y(1-y)
        layer_error1 = np.dot((self.output-self.y)*sigmoid_der(self.output),self.W1.T) #dJ/dh=dJ/dy*dy/dh
        d_W0 = np.dot(self.input.T, layer_error1*sigmoid_der(self.hidden_output)) #dJ/dh*dh/dW0
        
        self.W0 = self.W0 - self.learning_rate*d_W0
        self.W1 = self.W1 - self.learning_rate*d_W1 
    
    def train(self):
        for i in range(self.iterations):
            self.forward()
            self.backpropagation()
            self.iteration.append(i)
            self.error_history.append(self.loss)
    
    def predict(self,new_data):
        hidden_output = sigmoid(np.dot(new_data, self.W0)) #sigmoid(w0*x)
        output = sigmoid(np.dot(hidden_output, self.W1)) #sigmoid(w1*h)
        return output
```

### Learning Rate and Layer Tests
Bit harder to do this iteratively giving we have to decide an optimal layer configuration and a learning rate, so we'll just test a couple manually

```python
# Configure the tests
layer_numbers = [2, 16, 1] # 12 Hidden Layers
learning_rate = 0.01 # Smaller Learning Rate
iterations = 10000
Net = NeuralNetwork(x_train, y_train, layer_numbers, learning_rate, iterations)
Net.train()
plt.figure()
plt.plot(Net.iteration, Net.error_history)
```

```python
layer_numbers = [2, 2, 1] # 2 Hidden Layers
learning_rate = 0.0001
iterations = 10000
Net = NeuralNetwork(x_train, y_train, layer_numbers, learning_rate, iterations)
Net.train()
plt.figure()
plt.plot(Net.iteration, Net.error_history)
```

```python
layer_numbers = [2, 12, 1] # 12 Hidden Layers
learning_rate = 0.001 # Larger Learning Rate 
iterations = 10000
Net = NeuralNetwork(x_train, y_train, layer_numbers, learning_rate, iterations)
Net.train()
plt.figure()
plt.plot(Net.iteration, Net.error_history)
```

### Analysis Of Tests
12 Hidden Layers with 0.001 learning rate results in a smaller error, however it compromises efficiency a little. 

I personally like the results of a 0.01 learning rate and 16 hidden layers, small error, with a little more efficiency.

So we'll move forward with that configuration.

```python
# full prediction
layer_numbers = [2, 16, 1] # 12 Hidden Layers
learning_rate = 0.01 # Smaller Learning Rate
iterations = 10000
Net = NeuralNetwork(x_train, y_train, layer_numbers, learning_rate, iterations)
Net.train()
plt.figure()
plt.plot(Net.iteration, Net.error_history)

y_predict = Net.predict(x_test)
y_prediction = y_predict*(y_max - y_min) + y_min
y_test = y_test*(y_max - y_min) + y_min
plt.scatter(y_test, y_prediction)
plt.xlim(y_min, y_max)
plt.ylim(y_min, y_max)
```

```python
RMSE(y_test, y_prediction)
```

## Neural Network with Pytorch

```python
import torch
torch.random.manual_seed(2)
```

```python
# Load the data
data = pd.read_csv('data.csv')
X = data.iloc[:,2:] # Use all sensing variables
Y = data.iloc[:, 1] # Get part thickness

# convert to np array
X = X.to_numpy()
Y = Y.to_numpy()
```

```python
# Data normalization
np.random.shuffle(X)
x_norm = np.zeros((X.shape[0], X.shape[1]))
for i in range(X.shape[1]):
    data_ = X[:, i]
    x_norm[:, i] = (data_-np.amin(data_))/(np.amax(data_)-np.amin(data_))
y_min = np.min(Y)
y_max = np.max(Y)
y_norm = (Y-y_min)/(y_max-y_min)
```

```python
# Data Splitting 70 training / 30 testing
m = 0.7 * X.shape[0]
m = np.floor(m)
m = int(m)
x_train = x_norm[:m, :]
y_train = y_norm[:m]
x_test = x_norm[m:, :]
y_test = y_norm[m:]
y_train = np.reshape(y_train,(len(y_train),1))
y_test = np.reshape(y_test,(len(y_test),1))
print(x_test.shape, y_test.shape)
```

```python
# Convert to torch tensor
x_train_tensor = torch.tensor(x_train).float()
y_train_tensor = torch.tensor(y_train).float()
x_test_tensor = torch.tensor(x_test).float()
y_test_tensor = torch.tensor(y_test).float()
```

```python
class Net(torch.nn.Module):
    def __init__(self, layer_numbers):
        super().__init__()
        self.hidden1 = torch.nn.Linear(layer_numbers[0],layer_numbers[1], bias=False) # 3-layer perceptron
        self.output = torch.nn.Linear(layer_numbers[1],layer_numbers[2], bias=False)
        self.sigmoid = torch.nn.Sigmoid()
    
    def forward(self, x):
        x = self.hidden1(x)
        x = self.sigmoid(x)
        x = self.output(x)
        x = self.sigmoid(x)
        return x
```

### Testing different layer configurations and learning rates

```python
layer_numbers = [8, 16, 1] # 3 layer perceptron 8 -> 16 -> 1
iterations = 10000
net = Net(layer_numbers)

testing_rate = 0.01 # Learning rate of 0.01

criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=testing_rate)
loss_history = np.zeros(iterations)

for iteration in range(iterations):
    y_prediction = net(x_train_tensor)
    loss = criterion(y_prediction, y_train_tensor)
    optimizer.zero_grad()
    loss.backward() # back-prop
    optimizer.step() # update weights
    loss_history[iteration] = loss # for plotting
plt.plot(np.arange(iterations),loss_history)
```

```python
layer_numbers = [8, 32, 1] # 3 layer perceptron 8 -> 32 -> 1
iterations = 10000
net = Net(layer_numbers)

testing_rate = 0.1 # Learning rate of 0.1

criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=testing_rate)
loss_history = np.zeros(iterations)

for iteration in range(iterations):
    y_prediction = net(x_train_tensor)
    loss = criterion(y_prediction, y_train_tensor)
    optimizer.zero_grad()
    loss.backward() # back-prop
    optimizer.step() # update weights
    loss_history[iteration] = loss # for plotting
plt.plot(np.arange(iterations),loss_history)
```

```python
layer_numbers = [8, 32, 1] # 3 layer perceptron 8 -> 32 -> 1
iterations = 10000
net = Net(layer_numbers)

testing_rate = 0.01 # Learning rate of 0.01

criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=testing_rate)
loss_history = np.zeros(iterations)

for iteration in range(iterations):
    y_prediction = net(x_train_tensor)
    loss = criterion(y_prediction, y_train_tensor)
    optimizer.zero_grad()
    loss.backward() # back-prop
    optimizer.step() # update weights
    loss_history[iteration] = loss # for plotting
plt.plot(np.arange(iterations),loss_history)
```

### Analysis
For these tests, the learning rate of 0.1 with a layer configurations of 8 -> 32 -> 1 really takes the cake but shows some off grainular behavior towards the end which is worry. So I'll go with the 3rd option of 0.01 LR with configuration of 8 -> 32 -> 1

```python
layer_numbers = [8, 32, 1] # 3 layer perceptron 8 -> 32 -> 1
iterations = 10000
net = Net(layer_numbers)

criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=0.01)
loss_history = np.zeros(iterations)

for iteration in range(iterations):
    y_prediction = net(x_train_tensor)
    loss = criterion(y_prediction, y_train_tensor)
    optimizer.zero_grad()
    loss.backward() # back-prop
    optimizer.step() # update weights
    loss_history[iteration] = loss # for plotting
```

```python
# prediction and RMSE error
y_predict = net(x_test_tensor)
y_predicted = y_predict.detach()*(y_max-y_min) + y_min
y_test = y_test*(y_max-y_min) + y_min
plt.scatter(y_test, y_predicted)
plt.xlim(y_min, y_max)
plt.ylim(y_min, y_max)
plt.xlabel("Actual Part Thickness")
plt.ylabel("Predicted Part Thickness")
RMSE(y_test, y_predicted.numpy())
```

## Bonus: 4-Layer Perceptron with Pytorch

We just have to add an extra hidden layer to our layer configuration from the previous example

```python
class Net4(torch.nn.Module):
    def __init__(self, layer_numbers):
        super().__init__()
        self.hidden1 = torch.nn.Linear(layer_numbers[0],layer_numbers[1], bias=False) # 4-layer perceptron
        self.hidden2= torch.nn.Linear(layer_numbers[1], layer_numbers[2], bias=False) # Add the extra layer
        self.output = torch.nn.Linear(layer_numbers[2],layer_numbers[3], bias=False)
        self.sigmoid = torch.nn.Sigmoid() # Activation Function
    
    def forward(self, x): # Connect extra layer to forward feedings
        x = self.hidden1(x)
        x = self.sigmoid(x)
        x = self.hidden2(x)
        x = self.sigmoid(x)
        x = self.output(x)
        x = self.sigmoid(x)
        return x
```

```python
# Load the data
data = pd.read_csv('data.csv')
X = data.iloc[:,3:6] # Use all sensing variables
Y = data.iloc[:, 1] # Get part thickness

# convert to np array
X = X.to_numpy()
Y = Y.to_numpy()

# Data normalization
np.random.shuffle(X)
x_norm = np.zeros((X.shape[0], X.shape[1]))
for i in range(X.shape[1]):
    data_ = X[:, i]
    x_norm[:, i] = (data_-np.amin(data_))/(np.amax(data_)-np.amin(data_))
y_min = np.min(Y)
y_max = np.max(Y)
y_norm = (Y-y_min)/(y_max-y_min)

# Data Splitting 70 training / 30 testing
m = 0.7 * X.shape[0]
m = np.floor(m)
m = int(m)
x_train = x_norm[:m, :]
y_train = y_norm[:m]
x_test = x_norm[m:, :]
y_test = y_norm[m:]
y_train = np.reshape(y_train,(len(y_train),1))
y_test = np.reshape(y_test,(len(y_test),1))
print(x_test.shape, y_test.shape)

# Convert to torch tensor
x_train_tensor = torch.tensor(x_train).float()
y_train_tensor = torch.tensor(y_train).float()
x_test_tensor = torch.tensor(x_test).float()
y_test_tensor = torch.tensor(y_test).float()

layer_numbers = [3, 6, 12, 1] # 4 layer perceptron 3 -> 6 -> 12 -> 1 (random configuration)
iterations = 10000
net = Net4(layer_numbers)

criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=0.001)
loss_history = np.zeros(iterations)

for iteration in range(iterations):
    y_prediction = net(x_train_tensor)
    loss = criterion(y_prediction, y_train_tensor)
    optimizer.zero_grad()
    loss.backward() # back-prop
    optimizer.step() # update weights
    loss_history[iteration] = loss # for plotting
    
# prediction and RMSE error
y_predict = net(x_test_tensor)
y_predicted = y_predict.detach()*(y_max-y_min) + y_min
y_test = y_test*(y_max-y_min) + y_min
plt.scatter(y_test, y_predicted)
plt.xlim(y_min, y_max)
plt.ylim(y_min, y_max)
plt.xlabel("Actual Part Thickness")
plt.ylabel("Predicted Part Thickness")
RMSE(y_test, y_predicted.numpy())
```
